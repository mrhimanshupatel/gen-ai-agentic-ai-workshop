```python
sentence = " I love machine learning"

sentece -> vector?

averge Word2Vec

first -> create a vector for each word in the sentence

then -> take the average of all the vectors to get a single vector representation for the entire sentence   
```

```python
sentence = " I loVe Machine leaRning"
```

```python
sentence.lower().split()
```

```python
words = sentence.lower().split()
```

```python
word_vectors = []
```

```python
for word in words:
    print(word)
    word_vectors.append(model[word])
    
```

```python
word_vectors
```

```python
import numpy as np
sentence_vector =  np.mean(word_vectors,axis=0)
```
