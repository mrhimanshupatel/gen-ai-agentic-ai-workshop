# Interactive Notebook Example
This `.ipynb` file demonstrates **interactive exploration** using the functions from `text_utils.py`

```python
# Import our reusable module
from text_utils import clean_text, count_words, get_text_stats
```

## Test with Different Examples
Let's try various text inputs and see the results inline:

```python
# Example 1: Simple text
text1 = "  Machine Learning is AMAZING!  "
print(f"Original: {repr(text1)}")
print(f"Cleaned: {clean_text(text1)}")
print(f"Words: {count_words(text1)}")
```

```python
# Example 2: Longer text with stats
text2 = "Artificial Intelligence and Natural Language Processing are transforming how we interact with technology."
stats = get_text_stats(text2)

# Display as a nice formatted output
for key, value in stats.items():
    print(f"{key:20s}: {value}")
```

```python
# Example 3: Visualize word lengths (if matplotlib was installed)
# This shows how notebooks excel at visualization
sample_texts = [
    "AI",
    "Deep Learning",
    "Natural Language Processing",
    "Large Language Models and Embeddings"
]

for text in sample_texts:
    stats = get_text_stats(text)
    print(f"{text:40s} -> {stats['word_count']} words, avg length: {stats['avg_word_length']:.2f}")
```
